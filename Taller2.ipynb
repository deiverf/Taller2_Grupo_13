```python
#Importar dependencias
import csv
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lectura de archivo
data = pd.read_csv("players_20.csv", delimiter=',')
```

```python
#Visualización
data
```

```python
# Identificar columnas con valores nulos y contar la cantidad de nulos por columna
nulos= data.isnull().sum()

print("Filas y columnas con valores faltantes")

# Imprimir el nombre de la columna y la cantidad de nulos por columna
for column, nulo in nulos.items():
    if nulo > 0:
        
        print(f"{column}: {nulo} ")
```

```python
#tipo de dato de cada columna
dtypes_dict = data.dtypes.to_dict()
for column, dtype in dtypes_dict.items():
    print(f"{column}: {dtype}")
```

```python
# Conteo de tipo de columnas 

text_columns = 0
numeric_columns = 0

for column, dtype in dtypes_dict.items():
    if dtype == "object":
        text_columns += 1
    else:
        numeric_columns += 1
print(f"Número de columnas de tipo texto: {text_columns}")
print(f"Número de columnas numéricas: {numeric_columns}")
```

```python
#Calcular el promedio de edad de tres clubes

# Seleccionar tres clubes de la data
equipos = ['Liverpool', 'Juventus', 'FC Barcelona']
df_equipos = data[data['club'].isin(equipos)]

# Calcular el promedio de edad para cada club
promedio_edad = df_equipos.groupby('club')['age'].mean().reset_index()

# Graficar 
sns.barplot(x='club', y='age', data=promedio_edad, palette='husl')
plt.title('Edad promedio de jugadores')
plt.xlabel('Club')
plt.ylabel('Edad promedio')
plt.show()
```

```python
#Calcular el promedio de estatura de tres clubes

# Seleccionar tres clubes de la data
equipos = ['Manchester City', 'Real Madrid', 'Deportivo Cali']
df_equipos = data[data['club'].isin(equipos)]

# Calcular el promedio de edad para cada club
promedio_estatura = df_equipos.groupby('club')['height_cm'].mean().reset_index()

# Graficar 
sns.barplot(x='club', y='height_cm', data=promedio_estatura, palette='husl')
plt.title('Estatura promedio de jugadores')
plt.xlabel('Club')
plt.ylabel('Estatura promedio')
plt.show()
```
